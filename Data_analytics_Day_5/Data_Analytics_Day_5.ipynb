```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df=pd.read_csv('train.csv')
df
```

```python
al=df[df['Survived']=='alive']
a=sns.countplot(x='Sex',hue='Survived',data=al)
for i in a.containers:
    a.bar_label(i)
plt.show()
```

```python
df['Survived']=df['Survived'].replace({1:'alive',0:'die'})
```

```python
df.head()
```

```python
al=df[df['Survived']=='alive']
a=sns.countplot(x='Pclass',hue='Survived',data=al)
for i in a.containers:
    a.bar_label(i)
plt.show()
```

âž¡ 1st class passengers had the highest chance of survival.

```python
al=df[df['Survived']=='alive']
a=sns.heatmap(x='Embarked',hue='Survived',data=al)
for i in a.containers:
    a.bar_label(i)
plt.show()
```

```python

a=sns.barplot(x='Embarked', y='Survived', data=df)
plt.title('Survival Rate by Port of Embarkation')
plt.show()
```

Passengers from Cherbourg (C) often had higher survival (many were wealthy).

```python
df['Familysize']=df['SibSp']+df['Parch']+1
df
```

        âž¡ Indicates whether the passenger was alone or with family.

```python
df['Isalone']=1
df
```

```python
df.loc[df['Familysize']>1,'Isalone']=0
df
```

         âž¡ Helps identify if traveling alone affected survival.

```python
df['Title'] = df['Name'].str.extract(' ([A-Za-z]+)\.', expand=False)
df['Title'] = df['Title'].replace(['Mlle','Ms'],'Miss')
df['Title'] = df['Title'].replace('Mme','Mrs')

# Simplify rare titles
rare_titles = df['Title'].value_counts()[df['Title'].value_counts() < 10].index
df['Title'] = df['Title'].replace(rare_titles, 'Rare')
df
```

  âž¡ Titles (Mr, Mrs, Miss, Master, etc.) are strong survival indicators.

```python
bins = [0, 12, 18, 35, 60, 80]
labels = ['Child', 'Teen', 'Young Adult', 'Adult', 'Senior']
df['AgeGroup'] = pd.cut(df['Age'], bins=bins, labels=labels)
a=df['Age'].mean()
df['Age']=df['Age'].fillna(int(a))
df['Age']=df['Age'].astype(int)
df
```

```python
df['FareCategory'] = pd.qcut(df['Fare'], 4, labels=['Low', 'Mid', 'High', 'Very High'])
df
```

```python

a=sns.barplot(x='Familysize',y='Survived',data=df)
for i in a.containers:
    a.bar_label(i)
plt.show()
```

```python
sns.barplot(x='Title', y='Survived', data=df)
plt.title('Survival Rate by Title')
plt.show()
```

```python
sns.barplot(x='AgeGroup', y='Survived', data=df)
plt.title('Survival Rate by Age Group')
plt.show()
```

```python
numeric_df = df.select_dtypes(include=['int64', 'float64'])
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap (After Feature Engineering)')
plt.show()
```

```python
df_encoded = pd.get_dummies(df, columns=['Sex', 'Embarked', 'Title', 'AgeGroup', 'FareCategory'], drop_first=True)
df_encoded
```
