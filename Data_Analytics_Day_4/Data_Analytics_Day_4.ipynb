```python
import pandas as pd
import seaborn as sns
df=pd.read_csv('train.csv')
df.head()
```

```python
import matplotlib.pyplot as plt
sns.countplot(x='Sex',data=df)
plt.title('gender')
plt.show()
```

```python
sns.countplot(x='Survived',hue='Sex',data=df)
plt.title('Survival Distribution')
plt.xticks([0,1],['Died','Alive'])
plt.show()
```

```python
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title('Survival by Gender')
plt.legend(labels=['Died', 'Survived'])
plt.show()
```

```python
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title('Survival by Gender')
plt.legend(labels=['Died','Survived'],loc='best')
plt.show()
```

```python
sns.histplot(df['Age'], bins=30, kde=True)
plt.title('Age Distribution of Passengers')
plt.show()
```

Understand how age is spread  identify if most passengers were adults or children.

```python
sns.boxplot(x='Fare',data=df)
plt.title('Fare Distribution with Outliers')
plt.show()
```

Reveals skewness  a few passengers paid extremely high fares

```python
numeric_df = df.select_dtypes(include=['float64', 'int64'])
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

Observe relationships  e.g., Fare  Pclass, SibSp  Parch.

```python
sns.scatterplot(x='Age', y='Fare', hue='Pclass',marker='o', data=df)
plt.title('Age vs Fare by Class')
plt.show()
```

See how age and fare vary across passenger classes.

```python
sns.pairplot(df[['Age', 'Fare', 'Pclass', 'Survived']], hue='Survived')
plt.show()
```
